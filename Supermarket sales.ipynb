```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('supermarket_sales - Sheet1.csv')
df
```

```python
df = df.rename(columns = {"Total":"Total Sales"})
df
```

```python
df.info()
```

```python
sns.barplot(data=df,x='Quantity',y='Tax 5%')
```

```python
sns.countplot(data= df, x='City', hue= 'Branch')
```

```python
sns.countplot(data=df,x='City',hue='Gender')
```

```python
x=df['Gender'].value_counts()
print(x)
```

```python
plt.pie(x,labels=['M','F'],autopct='%1.0f%%')
```

```python
 sns.countplot(x="Payment", hue = "Branch", data=df).set_title("Payment Channel by Branch") 
```

```python
sns.boxplot(x="Branch", y = "Rating" ,data =df).set_title("Ratings by Branch") 
```

```python
 sns.countplot(x="Gender", hue = "Payment", data=df).set_title("Payment by Different Gender") 
```

```python
Y=df.groupby('Product line')['Total Sales'].sum()
print(Y)
sns.lineplot(data = Y)
plt.rcParams["figure.figsize"]=(10,10)
plt.xticks(rotation=60)
```

```python
#Customer type
#Branch
#Product line

plt.figure(figsize=(12, 6))
sns.barplot(x='Product line', y='Total Sales', hue='Customer type', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxenplot(data=df,x='Quantity',y='Product line')
```

```python
# since total sales is equal to income of company
Total_sales = df.groupby('Branch')['Total Sales'].sum()
print(Total_sales)

Total_sales.plot(kind='barh', title = 'Total income of each branch ')
plt.rcParams["figure.figsize"]=(5,5)
```

```python
z = df.groupby('Branch')['gross income'].sum()
print(z)
```

```python
sns.lineplot(data = z,label="Product sold in different Branch")
plt.rcParams["figure.figsize"]=(5,5)
```

```python
 sns.scatterplot(data = df,x='cogs',y='Rating')
```
